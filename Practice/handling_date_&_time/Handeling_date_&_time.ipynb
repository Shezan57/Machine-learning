# working with Date

```python
# converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

Extract Year

```python
date['date_year'] = date['date'].dt.year

date.sample(5)
```

Extract Month

```python
date['date_month_no'] = date['date'].dt.month

date.head(5)
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

Extract Days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['date_day_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# day of week name
date['date_week_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend ?
date['date_is_weekend'] = np.where(date['date_week_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.drop(columns=['product_id','city_id','orders']).head()
```

Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head(10)
```

Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id','city_id','orders']).head(10)
```
